```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('CarPrice.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse_output=False)
```

```python
categorical_col=['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']
encoded=ohe.fit_transform(dataset[categorical_col])
dataset=dataset.drop(categorical_col,axis=1)
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(categorical_col))
dataset=pd.concat([dataset.reset_index(drop=True),encoded_df.reset_index(drop=True)],axis=1)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset=dataset.drop('car_ID',axis=True)
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
scaling_col=['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']
dataset[scaling_col]=sc.fit_transform(dataset[scaling_col])
```

```python
dataset
```

```python
x=dataset.drop('price',axis=1)
```

```python
x
```

```python
y=dataset['price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100,max_depth=15,max_features=17,min_samples_split=10,min_samples_leaf=5,random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
rmse=np.sqrt(mse)
print(rmse)
```
